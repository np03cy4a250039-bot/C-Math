***Variable*** = A container that holds data values.


```python
hrs = input("Enter Hours:")
h = float(hrs)
rate = input("Enter the rate per hour:")
r = float(rate)
if h <=40:
  pay= h * r
else:
  overtimepay= h-40
  pay = (40*r) + (overtimepay*r*1.5)

print("Gross pay:", pay)
```
